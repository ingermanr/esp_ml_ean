### Muestreo aleatorio simple

Seleccionamos una muestra con `.sample()`, donde n es el tamaño de la muestra a extraer, frac es la fracción del df que se debe seleccionar como muestra, random_state es la semilla.

```python
# obtener una muestra de 100 filas del df

df_sample_n = df.sample(n=100, random_state=123) # fijamos la semilla para que sea reproducible
```

```python
df_sample_n
```

```python
# obtener una muestra del 20% del df

df_sample_frac = df.sample(frac=0.2, random_state=123) # fijamos la semilla para que sea reproducible
```

```python
df_sample_frac
```

### Muestreo aleatorio sistemático

Obtenemos la sistematicidad que queremos establecer (k = cada cuantos conteos tomar la muestra)

```python
# tamaño de la muestra

sample_size = 100

# calcular k

k = int(df.shape[0] / sample_size)

print(k)
```

```python
# generar indices sistemáticos

indices = list(range(0, df.shape[0], k))
```

```python
# ajustar si el número de indices es menor que el tamaño de la muestra

if len(indices) < sample_size:
    indices.append(df.shape[0])
```

```python
# seleccionar las muestras sistemáticas con los indices generados

df_sample_sist = df.iloc[indices]

df_sample_sist
```

### Muestreo aleatorio estratificado

Una caracteristica de los datos me agrupa segmentos o estratos de los datos. Quiero tomar muestras iguales de los estratos para equilibrar las proporciones.

```python
# crear función para crear estratos

def categorizar_columna(columna: pd.Series, estrato: str = 'pop'):
    if estrato in columna.lower():
        return estrato
    else:
        return 'otro'
```

```python
# crear columna estrato

df['estrato'] = df['top genre'].apply(categorizar_columna)
```

```python
df['estrato'].value_counts()/df.shape[0]
```

```python
# crear estratos con scikit-learn

from sklearn.model_selection import StratifiedShuffleSplit
```

```python
# inicializar el objeto StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits=1, test_size=sample_size, random_state=123)

# generar los indices de entrenamiento y prueba

for train_index, test_index in split.split(df, df['estrato']):
    df_train = df.iloc[train_index]
    df_test = df.iloc[test_index]

df_test["estrato"].value_counts()
```

### Muestreo aleatorio por conglomerados

Escojo los conlgomerados que necesito para dividir el df inicialmente, para luego tomar la muestra aleatoria simple de los conglomerados y tomo todos los registros de los conglomerados seleccionados en el muestreo

```python
# seleccionar la columna de conglomerados a tomar, para este caso year

conglomerados = df['year'].unique()

conglomerados
```

```python
# seleccionar los conglomerados con muestreo aleatorio simple

conglomerados_muestra = pd.Series(conglomerados).sample(n=5, random_state=123)

conglomerados_muestra
```

```python
# seleccionar las filas que pertenecen a los conglomerados seleccionados

df_muestra = df[df['year'].isin(conglomerados_muestra)]

df_muestra
```
