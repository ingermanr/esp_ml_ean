### Práctica de visualziación

```python
# importar librerias

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# importar dataset 'tips'

tips = sns.load_dataset('tips')
```

```python
# mostrar las primeras 5 filas del dataset

print(tips.head())
```

```python
# transformar en dataframe

df = pd.DataFrame(tips)
```

```python
# mostrar las primeras 5 filas del dataframe

print(df.head())
```

```python
# graficar por genero la proporción de tips

df_tips_by_sex = df.groupby("sex")['tip'].agg('sum')/df['tip'].agg('sum')
```

```python
df_tips_by_sex.plot(kind='pie', color=['skyblue', 'pink'], title='Proporción de tips por género', autopct='%1.1f%%')
plt.legend(["Hombre", "Mujer"],loc='upper right')
plt.show()
```

```python
# graficar fumadores y no fumadores por genero, con gráfico de barras

df_tips_by_sex_smoker = df.groupby("sex")["smoker"].value_counts().unstack()
```

```python
df_tips_by_sex_smoker.plot(kind='bar', stacked=True, color=['skyblue', 'pink'], 
                           title='Fumadores y no fumadores por género');
```
