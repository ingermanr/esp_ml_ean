## **Ejercicio 1: empresa de seguros**

Un analista de datos en una empresa de seguros esta examinando la cantidad de reclamos mensuales de una pliza de seguro. La cantidad de reclamos sigue una distribucin normal con una media de 50 reclamos al mes y una desviacin estndar de 10 reclamos.


* Cul es la probabilidad de que el nmero de reclamos en un mes sea menor o igual a 40?
* Cul es la probabilidad de que el nmero de reclamos en un mes sea mayor o igual a 60?
* Cul es la probabilidad de que el nmero de reclamos en un mes est entre 45 y 55?
* Qu nmero de reclamos corresponde al percentil 38?
* Qu nmero de reclamos corresponde al percentil 79?

```python
from scipy.stats import norm

# parmetros de la distribucin normal
mu = 50   # media
sigma = 10 # desviacin estndar

# 1. probabilidad de que el nmero de reclamos sea menor o igual a 40
x1 = 40
probabilidad_a_izquierda = norm.cdf(x1, mu, sigma)
print(f"La probabilidad de que el nmero de reclamos sea menor o igual a {x1} es {probabilidad_a_izquierda:.4f}")

# 2. probabilidad de que el nmero de reclamos sea mayor o igual a 60
x2 = 60
probabilidad_a_derecha = 1 - norm.cdf(x2, mu, sigma)
print(f"La probabilidad de que el nmero de reclamos sea mayor o igual a {x2} es {probabilidad_a_derecha:.4f}")

# 3. probabilidad de que el nmero de reclamos est entre 45 y 55
x3_1 = 45
x3_2 = 55
probabilidad_entre_valores = norm.cdf(x3_2, mu, sigma) - norm.cdf(x3_1, mu, sigma)
print(f"La probabilidad de que el nmero de reclamos est entre {x3_1} y {x3_2} es {probabilidad_entre_valores:.4f}")

# 4. percentil a la izquierda (38)
percentil_38 = norm.ppf(0.38, mu, sigma)
print(f"El nmero de reclamos correspondiente al percentil 38 es {percentil_38:.2f}")

# 5. Percentil a la izquierda (79)
percentil_79 = norm.ppf(0.79, mu, sigma)
print(f"El nmero de reclamos correspondiente al percentil 79 es {percentil_79:.2f}")
```

## **Ejercicio 2: ventas**

Eres un gerente de ventas en una empresa de productos electrnicos, y ests analizando las ventas mensuales de uno de los productos ms vendidos de la compaa. Se sabe que el volumen de ventas mensuales (en miles de unidades) sigue una distribucin normal con una media de 80,000 unidades y una desviacin estndar de 12,000 unidades.

* Cul es la probabilidad de que en un mes se vendan menos de 65,000 unidades?
* Cul es la probabilidad de que en un mes se vendan ms de 100,000 unidades?
* Cul es la probabilidad de que en un mes las ventas estn entre 70,000 y 90,000 unidades?
* Cul es el nmero de unidades vendidas que corresponde al percentil 20 (es decir, el nmero de unidades por debajo del cual estn el 20% de las ventas mensuales en la distribucin de densidad de las ventas)?
* Cul es el nmero de unidades vendidas que corresponde al percentil 85 (es decir, el nmero de unidades mensuales vendidas que correponde al 85% en la distribucin de densidad de las ventas)?

```python
from scipy.stats import norm

# parmetros de la distribucin normal
mu = 80000    # media (unidades vendidas en miles)
sigma = 12000 # desviacin estndar (variacin en ventas)

# 1. probabilidad de que las ventas sean menores a 65,000 unidades
x1 = 65000
probabilidad_a_izquierda = norm.cdf(x1, mu, sigma)
print(f"La probabilidad de vender menos de {x1} unidades es {probabilidad_a_izquierda:.4f}")

# 2. probabilidad de que las ventas sean mayores a 100,000 unidades
x2 = 100000
probabilidad_a_derecha = 1 - norm.cdf(x2, mu, sigma)
print(f"La probabilidad de vender ms de {x2} unidades es {probabilidad_a_derecha:.4f}")

# 3. probabilidad de que las ventas estn entre 70,000 y 90,000 unidades
x3_1 = 70000
x3_2 = 90000
probabilidad_entre_valores = norm.cdf(x3_2, mu, sigma) - norm.cdf(x3_1, mu, sigma)
print(f"La probabilidad de que las ventas estn entre {x3_1} y {x3_2} unidades es {probabilidad_entre_valores:.4f}")

# 4. percentil a la izquierda (20%)
percentil_20 = norm.ppf(0.20, mu, sigma)
print(f"El nmero de unidades vendidas correspondiente al percentil 20 es {percentil_20:.2f}")

# 5. percentil a la izquierda (85%)
percentil_85 = norm.ppf(0.85, mu, sigma)
print(f"El nmero de unidades vendidas correspondiente al percentil 85 es {percentil_85:.2f}")
```
